# Day 5 – Controlling Text Generation

```python
from transformers import pipeline
generator = pipeline('text-generation', model='gpt2')
```

```python
prompt = "In a distant galaxy"
temps = [0.2, 0.7, 1.2]
for t in temps:
    out = generator(prompt, max_length=50, temperature=t, do_sample=True)
    print(f"temp={t}: {out[0]['generated_text']}\n")
```

```python
# Greedy vs sampling
print("Greedy:", generator(prompt, max_length=30, do_sample=False)[0]['generated_text'])
print("Sampling:", generator(prompt, max_length=30, do_sample=True, temperature=0.8)[0]['generated_text'])
```

```python
# Interactive with Gradio (run in a separate cell or script)
# !pip install gradio
import gradio as gr

def generate(prompt, temperature, top_p, repetition_penalty):
    out = generator(prompt, max_length=100, temperature=temperature, top_p=top_p, repetition_penalty=repetition_penalty, do_sample=True)
    return out[0]['generated_text']

iface = gr.Interface(fn=generate, 
                     inputs=["text", gr.Slider(0.1, 1.5, 0.7), gr.Slider(0.5, 1.0, 0.9), gr.Slider(1.0, 2.0, 1.0)],
                     outputs="text",
                     title="Text Generator")
# iface.launch()  # uncomment to run
```
