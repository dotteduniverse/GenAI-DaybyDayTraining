```
from transformers import pipeline
gen = pipeline('text-generation', model='gpt2')
prompt = "The best thing about AI is"
for temp in [0.3,0.7,1.1]:
    for top_p in [0.6,0.9]:
        out = gen(prompt, max_length=40, temperature=temp, top_p=top_p, do_sample=True)
        print(f"temp={temp}, top_p={top_p}: {out[0]['generated_text']}\n")
```
