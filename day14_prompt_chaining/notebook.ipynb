```
import openai
from dotenv import load_dotenv
import os
load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")

def ask(prompt, temp=0):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        temperature=temp
    )
    return response.choices[0].message.content
```

```
# Chain: summarise → keywords
long_text = """
The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.
It was constructed from 1887 to 1889 as the centerpiece of the 1889 World's Fair.
Designed by Gustave Eiffel, it is one of the most famous landmarks in the world.
"""

summary = ask(f"Summarize this in one sentence:\n{long_text}")
print("Summary:", summary)
keywords = ask(f"Extract 3 keywords from: {summary}")
print("Keywords:", keywords)
```

```
# Self-consistency for math
math_question = "Roger has 5 apples. He gives 2 to his friend and buys 3 more. How many does he have?"
answers = []
for _ in range(5):
    ans = ask(f"Answer step by step: {math_question}", temp=0.7)
    # Extract the final number (simplistic)
    import re
    numbers = re.findall(r'\d+', ans)
    if numbers:
        answers.append(int(numbers[-1]))
print("Answers:", answers)
from collections import Counter
majority = Counter(answers).most_common(1)[0][0]
print("Majority answer:", majority)
```

```
# Conditional chain
review = "The product is amazing! I love it."
sentiment = ask(f"Classify sentiment as POSITIVE or NEGATIVE: {review}")
if "POSITIVE" in sentiment:
    reply = ask("Write a thank-you message for a positive review.")
else:
    reply = ask("Write a message asking for more details to resolve the issue.")
print(reply)
```
