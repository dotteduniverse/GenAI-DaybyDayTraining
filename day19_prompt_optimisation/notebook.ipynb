```
import dspy
from dspy import Predict, ChainOfThought

# Set up OpenAI (or local) as LM
lm = dspy.OpenAI(model="gpt-3.5-turbo")
dspy.settings.configure(lm=lm)

# Define signature
class Sentiment(dspy.Signature):
    """Classify sentiment as positive or negative."""
    text = dspy.InputField()
    sentiment = dspy.OutputField(desc="positive or negative")

# Basic predictor
predict = Predict(Sentiment)
result = predict(text="I love this product!")
print(result.sentiment)
```

```
# Optimise with few-shot
trainset = [
    dspy.Example(text="Great movie!", sentiment="positive"),
    dspy.Example(text="Terrible service", sentiment="negative"),
]
from dspy.teleprompt import BootstrapFewShot
optimizer = BootstrapFewShot(metric=lambda x,y: x.sentiment == y.sentiment)
optimized = optimizer.compile(predict, trainset=trainset)
result2 = optimized(text="Awesome experience")
print(result2.sentiment)
```
