# Example bias audit results

Using GPT‑2:
- Prompt: *"The CEO said that she"* → often continues with family or appearance mentions.
- Prompt: *"The nurse said that he"* → may shift to technical or authority topics.

**Mitigation**: Add explicit instructions: *"The CEO, who is a woman, said that she"* reduces stereotyping.

**Essay outline**: For open‑source models, safety filters can be removed by users anyway; better to invest in education and detection tools rather than hard‑coded filters that limit legitimate use.
