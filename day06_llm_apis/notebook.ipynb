```
import openai
import cohere
from dotenv import load_dotenv
import os

load_dotenv()
openai.api_key = os.getenv('OPENAI_API_KEY')
co = cohere.Client(os.getenv('COHERE_API_KEY'))
```

```
# OpenAI Q&A
context = "The Eiffel Tower is located in Paris, France. It was built in 1889."
question = "When was the Eiffel Tower built?"
prompt = f"Answer based on context:\nContext: {context}\nQuestion: {question}\nAnswer:"
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": prompt}]
)
print(response.choices[0].message.content)
```

```
# Cohere Q&A
co_response = co.generate(
    model='command',
    prompt=prompt,
    max_tokens=50,
    temperature=0.3
)
print(co_response.generations[0].text)
```

```
# Simple interactive loop
while True:
    q = input("Ask a question (or quit): ")
    if q.lower() == 'quit':
        break
    full_prompt = f"Answer concisely: {q}"
    resp = openai.ChatCompletion.create(model="gpt-3.5-turbo", messages=[{"role": "user", "content": full_prompt}])
    print("Answer:", resp.choices[0].message.content)
```
