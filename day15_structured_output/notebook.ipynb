```
import openai
import json
from dotenv import load_dotenv
import os
load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")

def ask_json(prompt):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        response_format={"type": "json_object"}
    )
    return json.loads(response.choices[0].message.content)

text = "John is 30 years old and lives in New York."
prompt = f"Extract name, age, city as JSON. Text: {text}"
data = ask_json(prompt)
print(data)
```

```
# Function calling
functions = [
    {
        "name": "extract_person_info",
        "description": "Extract person information",
        "parameters": {
            "type": "object",
            "properties": {
                "name": {"type": "string"},
                "age": {"type": "integer"},
                "city": {"type": "string"}
            },
            "required": ["name", "age", "city"]
        }
    }
]

response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[{"role": "user", "content": "Maria is 25 and lives in London."}],
    functions=functions,
    function_call="auto"
)
args = json.loads(response.choices[0].message.function_call.arguments)
print(args)
```

```
# Batch extraction
texts = [
    "Alice, 28, Paris",
    "Bob (age 35) from Berlin",
    "Carol lives in Madrid, she is 40"
]
results = []
for t in texts:
    resp = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": f"Extract name, age, city as JSON: {t}"}],
        functions=functions,
        function_call={"name": "extract_person_info"}
    )
    args = json.loads(resp.choices[0].message.function_call.arguments)
    results.append(args)
print(results)
```
