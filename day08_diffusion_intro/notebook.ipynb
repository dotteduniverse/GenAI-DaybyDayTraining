```
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
```

```
prompt = "a fantasy landscape with mountains and a glowing lake, digital art"
image = pipe(prompt, guidance_scale=7.5).images[0]
image
```

```
# Try different guidance scales
for scale in [2, 7, 12]:
    img = pipe(prompt, guidance_scale=scale).images[0]
    img.save(f"guidance_{scale}.png")
    print(f"Saved guidance_{scale}.png")
```

```
# Negative prompt
negative = "blurry, ugly, distorted, low quality"
img = pipe(prompt, negative_prompt=negative, guidance_scale=7.5).images[0]
img
```
