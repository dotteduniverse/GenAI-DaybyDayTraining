# Day 3 – Tokenization & Embeddings

Learn how text is split into tokens and converted into semantic vectors.

## Part 1: Tokenization with tiktoken (OpenAI)

We'll use `tiktoken` to see how a sentence becomes a list of token IDs.

```python
# Install if needed
# !pip install tiktoken
```

```python
import tiktoken

# Load the tokenizer used by GPT-4 / GPT-3.5-turbo
enc = tiktoken.get_encoding("cl100k_base")

sentence = "Tokenization is fun! 🎉"
tokens = enc.encode(sentence)
print("Tokens:", tokens)
print("Decoded:", enc.decode(tokens))
print("Number of tokens:", len(tokens))
```

```python
# Compare token counts across languages
english = "The quick brown fox jumps over the lazy dog."
german = "Der schnelle braune Fuchs springt über den faulen Hund."
chinese = "快速的棕色狐狸跳过懒狗。"

for text in [english, german, chinese]:
    tokens = enc.encode(text)
    print(f"{text[:30]}... -> {len(tokens)} tokens")
```

## Part 2: Visualizing Token Boundaries with Hugging Face

We'll use a BERT tokenizer to highlight how a sentence is split.

```python
# !pip install transformers
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
text = "I love natural language processing!"
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)
print("Token IDs:", tokenizer.convert_tokens_to_ids(tokens))
```

## Part 3: Generating Embeddings

We'll use a lightweight sentence‑transformer to turn sentences into vectors.

```python
# !pip install sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = [
    "A cat sits on a mat.",
    "A dog plays in the park.",
    "A feline rests on a rug.",
    "The stock market crashed today."
]

embeddings = model.encode(sentences)
print("Shape of embeddings:", embeddings.shape)  # (4, 384)
```

## Part 4: Cosine Similarity

Compare how similar sentences are in meaning.

```python
from sklearn.metrics.pairwise import cosine_similarity

sim_matrix = cosine_similarity(embeddings)
print("Cosine similarity matrix:")
print(np.round(sim_matrix, 2))
```

```python
# Find the most similar pair
for i in range(len(sentences)):
    for j in range(i+1, len(sentences)):
        print(f"{sentences[i][:20]} vs {sentences[j][:20]}: {sim_matrix[i][j]:.2f}")
```
