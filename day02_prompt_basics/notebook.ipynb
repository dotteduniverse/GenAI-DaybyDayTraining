## Setup
We'll use OpenAI's GPT-3.5-turbo. Make sure your API key is set in a `.env` file or as an environment variable.

```python
# Install if needed
# !pip install openai python-dotenv
```

```python
import openai
from dotenv import load_dotenv
import os

# Load API key from .env file
load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")

# If you don't have a .env file, set it directly (not recommended for sharing)
# openai.api_key = "your-key-here"
```

```python
# Helper function to call the LLM
def ask_gpt(prompt, temperature=0.7, max_tokens=200):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        temperature=temperature,
        max_tokens=max_tokens
    )
    return response.choices[0].message.content.strip()
```

## Your 10 Prompts

Fill in the `prompts` list below with 10 different prompts. Then run the cell to see the model's responses.

```python
# Write your 10 prompts here
prompts = [
    "Translate 'hello' to Spanish.",
    "You are a travel agent. Recommend a 3-day itinerary for Paris.",
    "Classify sentiment of these reviews: 'I love this!' -> Positive; 'It's okay' -> Neutral; 'Terrible' -> Negative. Now classify: 'Absolutely fantastic!'",
    "Solve: 24 + 35 = ? Let's think step by step.",
    "Write a haiku about autumn.",
    "Describe a tree in exactly 50 words.",
    "Explain the difference between supervised and unsupervised learning.",
    "Do NOT use any technical jargon. Explain what a CPU does.",
    "What are three possible futures of renewable energy?",
    "I think 7*8=54. Is that correct? If not, explain the mistake."
]

# Store responses
responses = []

for i, p in enumerate(prompts, 1):
    print(f"\n--- Prompt {i} ---\n{p}\n")
    resp = ask_gpt(p, temperature=0.7)
    print(f"Response:\n{resp}\n")
    responses.append((p, resp))
```

## Experiment with Temperature

Take one prompt (e.g., the haiku prompt) and run it with different temperatures.

```python
test_prompt = "Write a haiku about autumn."
temperatures = [0.2, 0.7, 1.2]

for temp in temperatures:
    resp = ask_gpt(test_prompt, temperature=temp)
    print(f"\n--- temperature = {temp} ---")
    print(resp)
```
