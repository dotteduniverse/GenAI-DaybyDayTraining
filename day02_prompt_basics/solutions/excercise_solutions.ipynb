## Exercise 1: Vague → Precise

Original vague prompt: `"Tell me about history."`

Three precise rewrites:
1. *"What were the main causes of World War I?"*
2. *"Describe the daily life of a peasant in medieval Europe."*
3. *"How did the invention of the printing press change European society?"*

Running them through GPT-3.5-turbo:

````python
import openai
from dotenv import load_dotenv
import os

load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")

def ask(prompt, temp=0.7):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        temperature=temp
    )
    return response.choices[0].message.content.strip()

precise_prompts = [
    "What were the main causes of World War I?",
    "Describe the daily life of a peasant in medieval Europe.",
    "How did the invention of the printing press change European society?"
]

for p in precise_prompts:
    print(f"\n--- {p} ---")
    print(ask(p))
````

**Observation:** The precise prompts yield focused, detailed answers, while the vague prompt gives a broad, shallow overview.

## Exercise 2: Few‑shot learning for a new task

Teaching the model to convert casual text to formal email.

````python
few_shot_prompt = """
Convert the following casual messages into formal business emails.

Example 1:
Casual: "Hey, can you send me the Q3 report? Thanks!"
Formal: "Dear Team, Could you please share the Q3 report at your earliest convenience? Thank you."

Example 2:
Casual: "Meeting tomorrow at 3? Let's discuss the budget."
Formal: "Dear colleagues, This is a reminder of our meeting scheduled for tomorrow at 3 PM to discuss the budget. I look forward to our discussion."

Now convert this:
Casual: "Hey John, can you send me the report by Friday? Thx!"
Formal:
"""

print(ask(few_shot_prompt))
````

**Expected output:** A polite email requesting the report by Friday, with proper salutation and closing.

## Exercise 3: Negative constraints

Explain machine learning without using the words *algorithm*, *data*, or *model*.

````python
negative_prompt = """
Explain what machine learning is. You are NOT allowed to use the words "algorithm", "data", or "model".
Use simple, everyday language.
"""

print(ask(negative_prompt, temp=0.5))
````

**Sample successful answer (paraphrased):**
*“Machine learning is a way for computers to learn from examples, just like how a child learns to recognise cats by seeing many pictures. Instead of being told every rule, the computer finds patterns on its own and gets better with practice.”*

If the model still uses forbidden words, refine the prompt with stronger constraints: *“If you use any of those words, start over.”*

## Exercise 5: Temperature sweep

Prompt: *"Write a short story about a talking cat."*

Results at different temperatures (simulated for brevity; learners should run themselves):

- **0.0** – Very predictable, repetitive. Story is safe and generic.
- **0.3** – Coherent, some nice details.
- **0.6** – Creative and still logical. Good balance.
- **0.9** – More surprising, occasional off‑topic sentences.
- **1.2** – Often nonsensical or broken grammar.

**Conclusion:** For creative writing, 0.6–0.8 is ideal. For factual answers, use 0.0–0.3.
