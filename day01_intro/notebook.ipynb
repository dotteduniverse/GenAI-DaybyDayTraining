## Part 1: Text Generation using GPT‑2

We'll use the `transformers` library to load a small GPT‑2 model and generate text continuations.

```python
# Install required packages (run only once in Colab)
# !pip install transformers torch
```

```python
from transformers import pipeline

# Load the text generation pipeline with GPT-2
generator = pipeline('text-generation', model='gpt2')
```

```python
# Generate text from a prompt
prompt = "The future of artificial intelligence is"
output = generator(prompt, max_length=50, num_return_sequences=1, temperature=0.8)
print(output[0]['generated_text'])
```

### Experiment with different parameters
- **temperature** (0.1 = very deterministic, 1.5 = very random)
- **top_k** (only sample from the k most likely next words)

```python
# Try different temperatures
for temp in [0.3, 0.9, 1.5]:
    output = generator(prompt, max_length=50, temperature=temp, do_sample=True)
    print(f"\n--- temperature = {temp} ---")
    print(output[0]['generated_text'])
```

## Part 2: Image Generation with Stable Diffusion

We'll use the `diffusers` library to generate images from text prompts.

> **Note**: The first run will download the model (~2GB). On Colab, enable GPU (Runtime → Change runtime type → T4 GPU).

```python
# !pip install diffusers accelerate
```

```python
from diffusers import StableDiffusionPipeline
import torch

# Load the pipeline (use fp16 for speed on GPU)
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Generate an image
prompt = "a cat wearing a space helmet, digital art, realistic"
image = pipe(prompt, guidance_scale=7.5).images[0]
image
```

```python
# Try a different guidance scale (lower = more creative, higher = more faithful to prompt)
image_low_guidance = pipe(prompt, guidance_scale=3.0).images[0]
image_low_guidance
```
